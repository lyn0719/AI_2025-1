## 드라이브에서 csv 파일 불러오고 읽기

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2025-1_AI/diabetes.csv", header=None)
df
```

## 0번째 행을 컬럼명으로 설정

> iloc

```python
# 0번째 행을 컬럼명으로 설정
df.columns = df.iloc[0]  # 첫 번째 행을 컬럼명으로 사용
df = df[1:].reset_index(drop=True)  # 첫 번째 행 삭제하고 인덱스 초기화

df
```

## 결측치 확인

```python
df.isnull().sum()
```

# Outcome 컬럼 제거

```python
df = df.drop('Outcome', axis=1)
df
```

## 데이터 분리

```python
# 데이터와 타겟 분리
x = df.drop('BMI', axis=1).values
y = df['BMI'].values
```

## 데이터 정규화 및 트레인/테스트 데이터 분리

```python
# 정규화
scaler = StandardScaler()
x = scaler.fit_transform(x)

# 트레인 / 테스트 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## shape 확인

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

# 회귀 모델 생성

## 라이브러리 import

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

## SVR 모델 생성

```python
from sklearn.svm import SVR # 라이브러리 import
from sklearn.metrics import mean_squared_error

clf_svm = SVR(kernel='linear') # SVR 모델 생성
clf_svm.fit(x_train, y_train) # 모델 학습

pred_svm = clf_svm.predict(x_test) # 테스트 데이터 예측

# 평가 출력 (평균 제곱 오차 : MSE)
print("\n--- SVR (Support Vector Regression) ---")
# print("Predicted values:", pred_svm) 예측 값 출력 코드
print("Mean Squared Error:", mean_squared_error(y_test, pred_svm)) # 평균제곱오차
```

## 선형회귀(LR) 모델 생성

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

clf_lr = LinearRegression()
clf_lr.fit(x_train, y_train)

pred_lr = clf_lr.predict(x_test)

mse = mean_squared_error(y_test, pred_lr)

print("\n--- LR (LinearRegression) ---")
print("Mean Squared Error (MSE):", mse)
```

## 결정 트리(DT) 회귀 모델 생성

```python
from sklearn.tree import DecisionTreeRegressor

clf_dt = DecisionTreeRegressor(random_state=0)
clf_dt.fit(x_train, y_train)

pred_dt = clf_dt.predict(x_test)

mse = mean_squared_error(y_test, pred_dt)

print("\n--- DT (Decision Tree Regressor) ---")
print("Mean Squared Error (MSE):", mse)
```

## 랜덤 포레스트(RF) 회귀 모델 생성

```python
from sklearn.ensemble import RandomForestRegressor

rf_clf = RandomForestRegressor(random_state=0)
rf_clf.fit(x_train, y_train)
pred_rf = rf_clf.predict(x_test)

mse = mean_squared_error(y_test, pred_rf)

print("\n--- RF (RandomForestRegressor) ---")
print("Mean Squared Error (MSE):", mse)
```

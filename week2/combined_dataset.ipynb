```python
# 깃허브 주소에서 엑셀 파일을 불러와 csv 파일로 변환 후 불러오기

import pandas as pd
import io
import requests

# GitHub의 raw 엑셀 파일 URL
file_url = "https://github.com/MyungKyuYi/AI-class/raw/refs/heads/main/combined_dataset-1.xlsx"

# 엑셀 파일을 직접 불러오기 (requests 사용)
response = requests.get(file_url)
response.raise_for_status()  # 요청이 실패하면 에러 발생

# 메모리에서 엑셀 파일을 판다스 데이터프레임으로 변환
df = pd.read_excel(io.BytesIO(response.content), engine="openpyxl")

# CSV 형식으로 변환 (메모리에서 처리)
csv_buffer = io.StringIO()
df.to_csv(csv_buffer, index=False, encoding="utf-8-sig")

# 변환된 CSV 데이터를 다시 읽기
csv_buffer.seek(0)  # StringIO 객체의 처음 위치로 이동
df_csv = pd.read_csv(csv_buffer, low_memory=False)

# 데이터 확인
print(df_csv.head())
print(df_csv.columns)
```

```python
# 로컬 파일의 엑셀 파일을 csv 파일로 변환 후 불러오기

import pandas as pd
import io

# 엑셀 파일 경로
file_path = "C:\\workplace\\AI\\AI_2025-1\\week2\\combined_dataset-1.xlsx"

# 엑셀 파일 읽기 (`engine='openpyxl'` 사용)
df = pd.read_excel(file_path, engine="openpyxl")

# CSV로 변환하여 메모리에서 처리 (로컬 파일 저장 없이)
csv_buffer = io.StringIO()  # 메모리 내 StringIO 버퍼 생성
df.to_csv(csv_buffer, index=False, encoding="utf-8-sig")

# 변환된 CSV 데이터를 다시 읽기 (메모리에서 바로 읽기)
csv_buffer.seek(0)  # StringIO 객체의 처음 위치로 이동
df_csv = pd.read_csv(csv_buffer, low_memory=False)

# 데이터 확인
print(df_csv.head())  # 첫 5행 출력
print(df_csv.columns)  # 데이터 컬럼 출력
```
